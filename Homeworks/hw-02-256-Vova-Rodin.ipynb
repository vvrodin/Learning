```python
import numpy as np
import pandas as pd
from scipy import stats
import random
import matplotlib.pyplot as plt
import seaborn as sns
import itertools as itertools

df = pd.read_csv('groceries.csv', sep=',')
df.columns = ['id', 'fielddate', 'product']

#пункт а
print('пункт а')
print()

print('Количество уникальных заказов:', len(df['id'].unique()))
print()

df_best_sold = pd.DataFrame(df['product'].value_counts()).reset_index() # массив всех покупаемых продуктов, отсортированный по количеству продаж
plt.bar(df_best_sold['index'][0:10].values, df_best_sold['product'][0:10].values, width=0.7, alpha=0.2, color='blue')
plt.ylabel('Количество')
plt.xlabel('Продукты')
plt.title('Наиболее покупаемые продукты')
plt.show()

from wordcloud import WordCloud

df_best_sold = pd.DataFrame(df_best_sold)

df_best_sold['str'] = (' ' + df_best_sold['index'] + ' ')*df_best_sold['product']
counter = ''
for i in range(len(df_best_sold.index)):
        counter += df_best_sold['str'][i]
text = counter

wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()

#пункт б
print('пункт б')
print()

df['Units'] = 1
table = pd.pivot_table(df, index='id', values='Units', columns='product', aggfunc=np.sum, fill_value=0)
print(table)
print()

corr_mat = table.corr()
print(corr_mat)
print()

sns.heatmap(corr_mat, annot=False, robust=True)
plt.title('Тепловая карта взаимосвязи продуктов в чеках')
plt.show()

print(corr_mat['domestic eggs'].sort_values())
print()

print('Сильнее всего с domestic eggs коррелируют meat spreads, а хуже всего - pet care')
print()

max = np.max(corr_mat.values[corr_mat.values != 1])
print('Максимальная корреляция составляет', max)
print(corr_mat[corr_mat.isin([max]).any(axis=1)])
print('Как мы можeм заметить, максимальная корреляция достигается между preservation products и soups')
print()
print('Корреляция пирсона показывает лишь линейную тесноту переменных и очень чувствительная к выбросам.')
print()

#пункт в
print('пункт в')
print()

table_m_corr = table.copy()
table_m_corr[table_m_corr > 1] = 1 # ограничем единицей количество товаров в чеке
table_sum = pd.DataFrame(np.sum(table_m_corr))

table_sum = table_sum[table_sum < 10] #убираем товары которые суммарно в чеках появляются меньше 10 раз
table_sum = table_sum.dropna() 

table_m_corr.drop(table_sum.index, inplace=True, axis=1)

m_corr = {'product_1': [], 'product_2': [], 'matthews_corrcoef': []}

combin = itertools.combinations(table_m_corr.columns, 2) #количество всех сочитаний

for i in combin:
    a, b = i
    m_corr['product_1'].append(a)
    m_corr['product_2'].append(b)
    df = pd.crosstab(table_m_corr[a], table_m_corr[b])
    m_corr['matthews_corrcoef'].append((df[0][0] * df[1][1] - df[1][0] * df[0][1]) /np.sqrt((df[0][0] + df[1][0]) * (df[0][0] + df[0][1]) * (df[1][1] + df[1][0]) * (df[1][1] + df[0][1])))

m_corr = pd.DataFrame(m_corr).sort_values('matthews_corrcoef', ascending=False)
print(m_corr)
print()
print('Ник: Vova.Vova.Rodin, id успешной посылки: 84817494')
print()

#пункт г
print('пункт г')
print()

egg_frq = (table['domestic eggs'] >= 1).mean() #среднее значение True среди всех чеков
milk_frq = (table['whole milk'] >= 1).mean()
mae_frq = ((table['whole milk'] >= 1) & (table['domestic eggs'] >= 1)).mean()
print('lift для whole milk и domestic eggs равен', mae_frq / (milk_frq * egg_frq))
print()

# пункт д
print('пункт д')
print()

table[table > 1] = 1 #ограничем единицей количество товаров в чеке
table_sum = pd.DataFrame(np.sum(table))
table_sum = table_sum[table_sum < 10] #убираем товары которые суммарно в чеках появляются меньше 10 раз
table_sum = table_sum.dropna()
table.drop(table_sum.index, inplace=True, axis=1)

combin = itertools.combinations(table_m_corr.columns, 2) #количество всех сочитаний

lift = {'product_1': [], 'product_2': [], 'lift': []}

for i in combin:
    a, b = i
    lift['product_1'].append(a)
    lift['product_2'].append(b)
    a_frq = (table[a] >= 1).mean()
    b_frq = (table[b] >= 1).mean()
    ab_frq = ((table[a] >= 1) & (table[b] >= 1)).mean()
    lift['lift'].append(ab_frq/(a_frq * b_frq))

lift = pd.DataFrame(lift).sort_values('lift', ascending=False)
print(lift)

print('Ник:Vova.Vova.Rodin, id успешной посылки: 84825051')
```
